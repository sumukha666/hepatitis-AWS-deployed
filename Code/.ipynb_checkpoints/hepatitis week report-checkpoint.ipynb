```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.metrics import roc_curve, auc, accuracy_score
from sklearn.metrics import precision_recall_curve, confusion_matrix

from sklearn.preprocessing import Imputer
```

```python
df = pd.read_csv("../HCV-Egy-Data.csv")
df.head()
```

```python
features = list(df.columns)
#exclude_list = ['Baselinehistological staging','WBC','RBC','HGB','Plat','ALT 48','ALT after 24 w','RNA Base','RNA 4','RNA 12','RNA EOT','RNA EF','Baseline histological Grading','Age','Gender','BMI','Fever','Nausea/Vomting','Headache','Diarrhea','Fatigue & generalized bone ache','Jaundice','Epigastric pain Epigastric pain']
#print features
exclude_list = ['Baselinehistological staging','Age ','Gender','BMI','Fever','Nausea/Vomting','Headache ' ,'Diarrhea ','Fatigue & generalized bone ache ','Jaundice ' ,'Epigastric pain ', 'WBC','RBC','HGB','Plat','ALT 48','ALT after 24 w','RNA Base','RNA 4','RNA 12','RNA EOT','RNA EF','Baseline histological Grading','AST 1','ALT 48']
for i in exclude_list:
    #print i
    features.remove(i)

    
X = df[features]
y = df['ALT 36'] 


#weeks = list('ALT 1','ALT4','ALT 12','ALT 24','ALT 36')
#X= df[weeks]
print X.head
print y.size
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression 

classifier = LogisticRegression(random_state = 0) 

classifier.fit(X_train, y_train) 

y_pred = classifier.predict(X_test) 

print ("Accuracy : ", metrics.accuracy_score(y_test, y_pred))
```
